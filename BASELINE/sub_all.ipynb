```python
import pandas as pd
import os
from scipy.stats import mode

sub_path = '../result/'
# sub_list = os.listdir(sub_path)
sub_list = ['20230516_121017', '20230516_222825', '20230518_232630', '20230519_004356', '20230519_012220', '20230519_031010', '20230519_124821']
path_list = [os.path.join(sub_path, i) for i in sub_list if not i.startswith('.')]

data_frames = {}

for i, item in enumerate(sub_list):
    name = f'sub{i}'
    data_frames[name] = pd.read_csv(os.path.join(sub_path, item, 'submit.csv'))

# Concatenate the data frames
df_concat = pd.concat(data_frames.values(), axis=1)

# Calculate the ensemble result using mode for the 'label' column
ensemble_label = mode(df_concat['label'], axis=1)[0].ravel()

# len(ensemble_label)
# # Reshape the ensemble_label to a 1-dimensional array
# ensemble_label = ensemble_label.reshape(-1)

# # Create a new data frame for the ensemble result
df_ensemble = pd.DataFrame({'id': data_frames['sub0']['id'], 'label': ensemble_label})

# # Print the ensemble result
print("Ensemble Result:")
print(df_ensemble)
```

```python
df_ensemble.to_csv('../result/ensemble.csv', index=False)
```
